```python
#TEIC-module2 (Propensity score patching)

#enter the filepath2 before running the program

filepath1 = "../dataset/Propensity_score_file(TEIC-table15).csv" #filepath to the input data (TEIC-table 15)
filepath2 = "" #filepath to the folder where to save the result

import pandas as pd
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
from psmpy import PsmPy
from psmpy.functions import cohenD
from psmpy.plotting import *

data = pd.read_csv(filepath1) #load data
pd.options.display.max_rows = 1000

psm = PsmPy(data, treatment="expert", indx ="dummy_index",exclude = ["index","TDM"])
psm.logistic_ps(balance = True)
display(psm.predicted_data)
psm.knn_matched(matcher="propensity_logit", replacement=False, caliper=0.1)
psm.plot_match(Title="Side by side matched controls", Ylabel="Number ofpatients", Xlabel= "Propensity logit",names = ["expert", "non-expert"],save=True)


print("matched ID")
display(psm.matched_ids)
ls =[]
ls_ex =[]
ls_non_ex =[]
dummy_list1=[]
dummy_list2=[]
for id1, id2 in zip(psm.matched_ids["dummy_index"],psm.matched_ids["matched_ID"]):
    if len(data[data["dummy_index"]==id1].values.tolist())>0 and len(data[data["dummy_index"]==id2].values.tolist())>0:
        for i in data[data["dummy_index"]==id1].values.tolist()[0]:
            dummy_list1.append(i)
        ls_non_ex.append(dummy_list1)
        dummy_list1 =[]
        for j in data[data["dummy_index"]==id2].values.tolist()[0]:
            dummy_list2.append(j)
        ls_ex.append(dummy_list2)
        dummy_list2 =[]
    else:
        pass
    ls.append([id1,id2])

df_ls = pd.DataFrame(ls,columns=["dummy_index","matched ID"])
df_ex = pd.DataFrame(ls_ex,columns=data.columns)
df_non_ex =pd.DataFrame(ls_non_ex,columns=data.columns)
print("df_ex")
display(df_ex)
df_ex.to_csv(str(filepath2)+"/data_expert.csv")
print("df_non_ex")
display(df_non_ex)
df_non_ex.to_csv(str(filepath2)+"/data_non_expert.csv")
```
